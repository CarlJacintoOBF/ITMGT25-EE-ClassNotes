#### Understanding the data

**What is a Non-Performing Loan (NPL)?**

A nonperforming loan (NPL) is a sum of borrowed money upon which the debtor has not made the scheduled payments for a specified period. Although the exact elements of nonperformance status vary, depending on the specific loan's terms, "no payment" is usually defined as zero payments of either principal or interest. The specified period also varies, depending on the industry and the type of loan. Generally, however, the period is 90 days or 180 days.

Source: [Investopedia](https://www.investopedia.com/terms/n/nonperformingloan.asp)

#### 1.2. How can we convert the multi-level index column into a single column?

Discard the current multi-level column structure into a single column one.

Each column must follow this naming convention:
    
    Year-Month

Examples: 2008-March, 2008-June, ... , 2017-December

```python
## write code here



```

```python
## solution





## scroll down to see






new_columns = ["Category"]

current_year = '2008'

for i, (a,b) in enumerate(df.columns):
    # print(a,b)
    if i == 0:
        continue
    
    if( (a != " ") and (a != current_year)):
        current_year = a
    new_columns.append("-".join([current_year,b]))
    
new_columns
```

### 3. Data Visualization

#### 3.1 Draw a bar chart illustrating the trend in Motor Vehicle Non-Performing Loans as percentages of Total Non-Performing Loans. 

```python
## write your code here



```

```python
## Solution


## scroll down to see









import matplotlib.pyplot as plt

fig, ax = plt.subplots(1)
fig.set_figheight(8)
fig.set_figwidth(15)

label_font = {"fontweight":"bold", "fontfamily":"Lato", "fontsize":14}
title_font = {"fontfamily":"Lato", "fontsize":18}
suptitle_font = {"fontweight":"bold", "fontfamily":"Lato", "fontsize":24}

plt.suptitle("Non-Performing Motor Vehicle Loans to Total Non-Performing Loans (%)", y=1, **suptitle_font)
ax.bar(height=nonperf_mvl_total_npl_series,x=nonperf_mvl_total_npl_series.index.tolist(),color="Red")
ax.set_title("Source: Bangko Sentral ng Pilipinas", **title_font)
ax.set_xlabel("Quarter", **label_font)
ax.set_ylabel("Percentage (%)", **label_font)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
```
