Do that now.

### Queries

Find all rows where customers reside in Quezon City

```python
## scroll down to see solution







df.loc[df.loc[:,"city"]=="Quezon City"]
```

```python
# alternate solution using query
df.query("city=='Quezon City'")
```

Get the first names of males who live in Pasig

```python
## scroll to see solution










df.loc[(df.loc[:,"city"]=="Pasig") & (df.loc[:,"gender"]=="Male"),:"first name"]
```

```python
## alternate, more modern and shorter solution
df.query("city=='Pasig'").loc[:,"first name"]
```

Get the names (first name and last name) of those with cash balances greater than or equal to 10,000. Show the actual cash balance as well.

```python
## scroll to see solution







df.loc[df.loc[:,"cash balance"]>=10000,["first name","last name","cash balance"]]
```

Get the names (first name and last name) of those with points greater than or equal to 50,000 and who live in Pasay. Show the actual city and points as well.

```python
## scroll to see solution




df.loc[(df.loc[:,"points"]>=50000) & (df.loc[:,"city"]=="Pasay"),["first name","last name","city","points"]]
```

Get the names (first name and last name) of those with points greater than or equal to 50,000 but do not live in Quezon City. Show the actual city and points as well.

```python
## scroll to see solution



df.loc[(df.loc[:,"points"]>=50000) & (df.loc[:,"city"]!="Quezon City"),["first name","last name","city","points"]]

```

Get the names (first name and last name) of those with points greater than or equal to 50,000 but live in Quezon City or Mandaluyong. Show the actual city and points as well.

```python
## scroll to see solution






df.loc[(df.loc[:,"points"]>=50000) & (df.loc[:,"city"].isin(["Quezon City","Mandaluyong"])),["first name","last name","city","points"]]




```

```python
# alternate, old solution
```

```python
df.loc[(df.loc[:,"points"]>=50000) & ((df.loc[:,"city"]=="Quezon City") | (df.loc[:,"city"]=="Mandaluyong")),["first name","last name","city","points"]]

```

```python
# alternate, modern solution
df.query("points>=50000 & (city in ['Quezon City','Mandaluyong'])").loc[:,["first name","last name","city","points"]]
```
